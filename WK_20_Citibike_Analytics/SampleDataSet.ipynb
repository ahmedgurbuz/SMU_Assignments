## Creating a sample dataset for Citibike Analysis in Tableau

```python
#Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Files to Load
file01 = "201801-citibike-tripdata.csv"
file02 = "201802-citibike-tripdata.csv"
file03 = "201803-citibike-tripdata.csv"
file04 = "201804-citibike-tripdata.csv"
file05 = "201805-citibike-tripdata.csv" 
file06 = "201806-citibike-tripdata.csv"
file07 = "201807-citibike-tripdata.csv"
file08 = "201808-citibike-tripdata.csv"
file09 = "201809-citibike-tripdata.csv"
file10 = "201810-citibike-tripdata.csv"
file11 = "201811-citibike-tripdata.csv"
file12 = "201812-citibike-tripdata.csv"

# Read files and store them into Pandas data frame
jan = pd.read_csv(file01)
feb = pd.read_csv(file02)
mar = pd.read_csv(file03)
apr = pd.read_csv(file04)
may = pd.read_csv(file05)
jun = pd.read_csv(file06)
jul = pd.read_csv(file07)
aug = pd.read_csv(file08)
sep = pd.read_csv(file09)
octo = pd.read_csv(file10)
nov = pd.read_csv(file11)
dec = pd.read_csv(file12)
```

```python
# Get fraction of rows and store them
# https://www.geeksforgeeks.org/how-to-randomly-select-rows-from-pandas-dataframe/

jan18 = jan.sample(frac = 0.001)
feb18 = feb.sample(frac = 0.001)
mar18 = mar.sample(frac = 0.001)
apr18 = apr.sample(frac = 0.001)
may18 = may.sample(frac = 0.001)
jun18 = jun.sample(frac = 0.001)
jul18 = jul.sample(frac = 0.001)
aug18 = aug.sample(frac = 0.001)
sep18 = sep.sample(frac = 0.001)
octo18 = octo.sample(frac = 0.001)
nov18 = nov.sample(frac = 0.001)
dec18 = dec.sample(frac = 0.001)
```

```python
# Concatenate dataframes
# https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

year18 = jan18.append(feb18).append(mar18).append(apr18).append(may18).append(jun18).append(jul18)\
         .append(aug18).append(sep18).append(octo18).append(nov18).append(dec18)
```

```python
df_year18 = year18.reset_index(drop=True)
df_year18
```

```python
# Save dataframe to csv file
# https://stackoverflow.com/questions/26786960/remove-index-column-while-saving-csv-in-pandas
df_year18.to_csv('year18.csv',mode = 'w', index=False)
```
