### Creating age bins as a new column for Citibike Data

```python
# Import Dependencies
import pandas as pd
import numpy as np
import datetime as dt
from datetime import date

# File to Load
file = "year18.csv"
# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file)

df.head()
```

```python
df['age'] = 2019 - df['birth year']
```

```python
# Create bins in which to place values based upon Age Demographics
bins = [0, 9.99, 19.99, 29.99, 39.99, 49.99, 59.99, 69.99, 79.99, 200]

# Create labels for these bins
group = ["<10", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80+"]
```

```python
# Slice the data and place it into bins
# Place the data series into a new column inside of the DataFrame
df["ageGroup"] = pd.cut(df["age"], bins, labels = group)

df
```

```python
# Save dataframe to csv file
# https://stackoverflow.com/questions/26786960/remove-index-column-while-saving-csv-in-pandas
df.to_csv('year18.csv',mode = 'w', index=False)
```
